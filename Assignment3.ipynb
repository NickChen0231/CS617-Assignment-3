```python
sql = '''
SELECT "DISTRICT",
       "YEAR",
       "MONTH",
       "OCCURRED_ON_DATE",
       "SHOOTING",
       "OFFENSE_DESCRIPTION"
FROM "b973d8cb-eeb2-4e7e-99da-c92938efc9c0"
WHERE CAST("SHOOTING" AS INT)=1
    '''
```

```python
bpd_logs = "https://data.boston.gov/api/3/action/datastore_search_sql"
```

```python
import urllib, requests
```

```python
url = bpd_logs+"?sql="+urllib.parse.quote(sql)
rows = requests.get( url ).json()

shootings = rows['result']['records']
```

```python
shootings_per_year = [0, 0, 0, 0]

for s in shootings:
    if s['YEAR'] == '2023':
      shootings_per_year[0] += 1
    elif s['YEAR'] == '2024':
      shootings_per_year[1] += 1
    elif s['YEAR'] == '2025':
      shootings_per_year[2] += 1
    elif s['YEAR'] == '2026':
      shootings_per_year[3] += 1


shootings_per_year
```

```python
months_2023 = [0] * 12

for s in shootings:
    if s['YEAR'] == '2023':
        months_2023[int(s['MONTH']) - 1] += 1
months_2023
```
